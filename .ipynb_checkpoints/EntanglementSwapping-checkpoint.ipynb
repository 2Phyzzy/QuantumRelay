```python
"""ITU channel 34 will be used for the photons (1550.12nm)"""
print("frequency bandwidth is", sigma*(10**-9), "GHz")
"""For now assume 200fs laser wih 75MHz repetition rate, which gives
a Pulse Spacing"""

data = []
z2s = []
n = 2 #detectors

zmin = -0.01
zmax = 0.01
zstp = 0.0001

np.concatenate((data,np.array([0])))
z1 = 0
Omega1 = ITU_dict[ITU_centre][1]
Omega2 = ITU_dict[ITU_centre][1]
for z2 in np.arange(zmin,zmax,zstp):
    data1 = []
    z2s.append(z2)
    diss = []
    for d in np.arange(0,50.1,0.1):
        diss.append(d)        
        Pc = Pcoincidence_2bs(z1,z2,Omega1,Omega2,sigma)
        Pf = Pfibre(d,n)
        Ptot = Ptotal(Pf,None,Pc)
        data1.append(Ptot)
    data.append(data1)

data = np.array(data)
diss = np.array(diss)
z2s = np.array(z2s)

z2ss = z2s*1000
    
stuff = np.round(1000*np.arange(zmin,zmax,zstp),4)
diss = np.round(diss,3)
"""For data in dB"""
data = 10*np.log10(data)
data = pd.DataFrame(data,columns = diss,index = stuff)#z2ss)

plt.figure(figsize=(6,4))
ax = sns.heatmap(data, cbar_kws={'label': 'Probability of Coincidence Being Observed, dB'})#,cmap="YlGnBu")
plt.xlabel("fibre length, km")
plt.ylabel("path difference, mm")
plt.show()
```

### Now for the actual probability of the actual swap happening between Source 1 and Source 2

```python
"""Parameters"""
source_counts = 10**6
coincidence_window = 250*(10**-12)
dtot = 100
n = 4
source_rate = 10**6

S1 = 10*(10**6) #Hz
S2 = 10*(10**6) #Hz
S3 = 10*(10**6) #Hz
S4 = 10*(10**6) #Hz
```

```python
"""Tru Success"""
TS = True_Success(dtot,n,source_rate)#,z1=0,z2=0,sigma=100*(10**9),ITU1=34,ITU2=34)
FS = False_Success(dtot,n,S1,S2,S3,S4)#,coincidence_window=250*(10**-12))

print("Ture counts of ",TS*(10**(3)), "mHz, with error", FS*(10**(3)), "mHz")
```

```python
"""Sqrt of Alpha-Beta Coincidences"""

z1 = 0
z2 = 0
BSM_Sig = source_counts*Pcoincidence_2bs(0,0)*Pfibre(dtot/2,n/2)
BSM_Noise = np.sqrt(S1*Pcoincidence_2bs(0,0)*Pfibre(dtot/2,n/2))


print(BSM_Sig*(10**(-3)), "kHz, with error", BSM_Noise*(10**(-3)), "kHz")
```

```python
"""For a plot of Signal against individual fibre length"""
n = 4
source_rate = 10**6
S1 = 10*(10**6) #Hz
S2 = 10*(10**6) #Hz
S3 = 10*(10**6) #Hz
S4 = 10*(10**6) #Hz

TSs = []
FSs = []
BSM_N = []
BSM_S = []
dis = []
for i in np.arange(5,40.1,1):
    dtot = 4*i
    TS = True_Success(dtot,n,source_rate)
    FS = False_Success(dtot,n,S1,S2,S3,S4)
    BSM_Sig = (source_counts*Pfibre(2*i,2))*Pcoincidence_2bs(0,0)
    BSM_Noise = np.sqrt(((S1*Pfibre(2*i,2))**2)*Pcoincidence_2bs(0,0))
    dis.append(i)
    TSs.append(TS)
    FSs.append(FS)
    BSM_N.append(BSM_Noise)
    BSM_S.append(BSM_Sig)
    
TSs = np.array(TSs) 
FSs = np.array(FSs) 
BSM_S = np.array(BSM_S) 
BSM_N = np.array(BSM_N)
dis = np.array(dis)

plt.figure(figsize=(8,4))
plt.errorbar(dis,TSs,yerr=FSs,label="4-fold Coincidences",fmt='o-')
plt.errorbar(dis,BSM_S,yerr=BSM_N,label="Coincidences on Entanglement Swapping",fmt='o-')
plt.xlabel("Link length, km")
plt.ylabel("Coincidence Rate, 1/s")# (log10(1/s))")
plt.yscale('log')
plt.grid()

plt.legend()
plt.show()
    
```

```python
"""For a plot of Signal against total fibre length"""
n = 4
source_rate = 10**6
S1 = 10*(10**6) #Hz
S2 = 10*(10**6) #Hz
S3 = 10*(10**6) #Hz
S4 = 10*(10**6) #Hz

TSs = []
FSs = []
BSM_N = []
BSM_S = []
SL_T = []
SL_N = []
dis = []
for i in np.arange(5,100.1,5):
    dtot = i
    TS = True_Success(dtot,n,source_rate)
    FS = False_Success(dtot,n,S1,S2,S3,S4)
    BSM_Sig = source_counts*Pcoincidence_2bs(0,0)*Pfibre(dtot/2,2)
    BSM_Noise = np.sqrt(S1*Pcoincidence_2bs(0,0)*Pfibre(dtot/2,2))
    SL = source_counts*Pfibre(dtot,2)
    SLN = np.sqrt(S1*Pcoincidence_2bs(0,0)*Pfibre(dtot,2))
    dis.append(i)
    TSs.append(TS)
    FSs.append(FS)
    BSM_N.append(BSM_Noise)
    BSM_S.append(BSM_Sig)
    SL_T.append(SL)
    SL_N.append(SLN)
    
TSs = np.array(TSs) 
FSs = np.array(FSs) 
BSM_S = np.array(BSM_S) 
BSM_N = np.array(BSM_N)
SL_T = np.array(SL_T)
SL_N = np.array(SL_N)
dis = np.array(dis)

plt.figure(figsize=(8,4))
plt.errorbar(dis,TSs,yerr=FSs,label="4-fold Coincidences",fmt='-')
plt.errorbar(dis,BSM_S,yerr=BSM_N,label="Coincidences on Entanglement Swapping",fmt='-')
plt.errorbar(dis,SL_T,yerr=SL_N,label="Coincidences without Entanglement Swapping",fmt='-')
#plt.plot(dis,BSM_N,label="Noise on Entanglement Swapping")
#plt.plot(dis,BSM_S,label="True coincidences on Entanglement Swapping")
plt.xlabel("Link length, km")
plt.ylabel("Coincidence Rate, 1/s")# (log10(1/s))")
plt.yscale('log')
plt.grid()
#plt.xlim(4.99,5.01)
#plt.ylim(159500,161500)
plt.legend()
plt.show()
```
