# The probabilities of succesfully swapping entanglement

#### This document will calclate the probabilities and the rates of entanglement swapping for longdistance quantum communication.

Below are the definitions for the simulations and calculations.

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
import csv
import scipy
```

```python
ITU_dict = {17: [1563.86,191.7],18: [1563.05,191.8],19: [1562.23,191.9],20: [1561.42,192.0],
        21: [1560.61,192.1],22: [1559.79,192.2],23: [1558.98,192.3],24: [1558.17,192.4],
        25: [1557.36,192.5],26: [1556.55,192.6],27: [1555.75,192.7],28: [1554.94,192.8],
        29: [1554.13,192.9],30: [1553.33,193.0],31: [1552.52,193.1],32: [1551.72,193.2],
        33: [1550.92,193.3],34: [1550.12,193.4],35: [1549.32,193.5],36: [1548.51,193.6],
        37: [1547.72,193.7],38: [1546.92,193.8],39: [1546.12,193.9],40: [1545.32,194.0],
        41: [1544.53,194.1],42: [1543.73,194.2],43: [1542.92,194.3],44: [1542.12,194.4],
        45: [1541.35,194.5],46: [1540.56,194.6],47: [1539.77,194.7],48: [1538.98,194.8],
        49: [1538.19,194.9],50: [1537.40,195.0],51: [1536.61,195.1],52: [1535.82,195.2],
        53: [1535.04,195.3],54: [1534.25,195.4],55: [1533.47,195.5],56: [1532.68,195.6],
        57: [1531.90,195.7],58: [1531.12,195.8],59: [1530.33,195.9],60: [1529.55,196.0],
        61: [1528.77,196.1]}

def Pcoincidence(z1,z2,Omega1=ITU_dict[34][1],Omega2=ITU_dict[34][1],sig= 100*(10**9)):
    """This takes in 5 values and computes the probability of coincidences being seen in
    the Bell State Measurement, using the Coalescence Interference method.
    z1 is the path length of input 1, and Omega1 is the central frequency of that input.
    z2 is the path length of input 2, and Omega2 is the central frequency of that input.
    sig is the parameter for coherence time."""
    
    dz = abs(z2 - z1)                #m
    dOmega = abs(Omega2 - Omega1)*(10**12)    #1/s
    c = 299792458               #m/s
    Pc = 0.5*( 1 - np.exp(-0.5*(dz**2)*((sig/c)**2))*np.exp(-0.5*((dOmega/sig)**2)) )
    
    return Pc 

def Pcoincidence_2bs(z1,z2,Omega1=ITU_dict[34][1],Omega2=ITU_dict[34][1],sig= 100*(10**9)):
    """This takes in 5 values and computes the probability of coincidences being seen in
    the Bell State Measurement, using the Coalescence Interference method.
    z1 is the path length of input 1, and Omega1 is the central frequency of that input.
    z2 is the path length of input 2, and Omega2 is the central frequency of that input.
    sig is the parameter for coherence time."""
    
    dz = abs(z2 - z1)                #m
    dOmega = abs(Omega2 - Omega1)*(10**12)    #1/s
    c = 299792458               #m/s
    Pc = 0.25*( 1 + np.exp(-0.5*(dz**2)*((sig/c)**2))*np.exp(-0.5*((dOmega/sig)**2)) )
    
    return Pc 

def Pfibre(dtot,n,eta=0.8):
    """This takes in the total length of optical fibre (dtot) and the total number of APDs in
    the system (n) to give the probability of a photon being registered in all n detectors."""
    
    alpha = 0.3                 #dB/km
    #eta is detector efficiency for nanowire detectors
    
    t = 10**(-(alpha*dtot)/10)  #The loss in dtot length of fibre
    Pf = t*(eta**n)             #The total systematic loss including the fibre and the detector efficiency
    
    return Pf

def Ptotal(Pf,Pc=None,Pc_2bs=None):
    """This takes in the previous two probabilities and calculates the total probability of
    the entanglement being swapped abd a coincidence beins seen between the two users."""
    if Pc is None:
        Ptot = Pc_2bs*Pf
    if Pc_2bs is None:
        Ptot = Pf*(1-Pc)
        
    return Ptot


def find_nearest(array, value):
    array = np.asarray(array)
    idx = (np.abs(array - value)).argmin()
    return array[idx]

def PSource_SPDCalc(offset,sigma= 100*(10**9),plot=False):

    c = 299792458 #m/s
    ITU_centre = 34
    #offset is the ITU channel offset from 34
    ch1 = ITU_centre - offset
    l1 = ITU_dict[ch1][0]
    lamda1 = l1*(10**(-9))
    dlamda1 = (sigma*(lamda1**2)/c)*(10**9)
    lmin1 = (l1 - (dlamda1/2))
    lmax1 = (l1 + (dlamda1/2))
    ch2 = ITU_centre + offset
    l2 = ITU_dict[ch2][0]
    lamda2 = l2*(10**(-9))
    dlamda2 = (sigma*(lamda2**2)/c)*(10**9)
    lmin2 = (l2 - (dlamda2/2))
    lmax2 = (l2 + (dlamda2/2))
        
    data = np.genfromtxt('Laser_Data\spdcalc_data_large_trim.csv', delimiter=',')
    sig = data[0,1:]
    ide = data[1:,0]
    data = data[1:,1:]
    df = pd.DataFrame(data, index = np.round(ide,1), columns = np.round(sig,1))
    
    sigmin = find_nearest(sig,lmin1)
    sigminloc = np.where(sig == sigmin)[0][0]
    sigmax = find_nearest(sig,lmax1)
    sigmaxloc = np.where(sig == sigmax)[0][0]
    idemin = find_nearest(ide,lmin2)
    ideminloc = np.where(ide == idemin)[0][0]
    idemax = find_nearest(ide,lmax2)
    idemaxloc = np.where(ide == idemax)[0][0]
    
    """The slice to see all photons in the Signal DWDM"""
    cutdata = data[:,sigminloc:sigmaxloc]
    cutsig  = sig[sigminloc:sigmaxloc]
    cutide  = ide
    dataSig = cutdata
    dataSig_df = pd.DataFrame(cutdata, columns = cutsig, index = cutide)
    
    totSig = 0
    for i in range(len(dataSig)):
        
        for j in range(len(dataSig[i])):
            totSig += cutdata[i][j]
    
    
    """The slice to see all photons in the Idler DWDM"""
    cutdata = data[idemaxloc:ideminloc,:]
    cutsig  = sig
    cutide  = ide[idemaxloc:ideminloc]
    dataIde = cutdata
    dataIde_df = pd.DataFrame(cutdata, columns = np.round(cutsig,4), index = np.round(cutide,4))
    
    totIde = 0
    for i in range(len(dataIde)):
        
        for j in range(len(dataIde[i])):
            totIde += cutdata[i][j]
    
    
    """The slice that shows only the possible entangled photons"""
    cutdata = data[idemaxloc:ideminloc,sigminloc:sigmaxloc]
    #print("data =",cutdata)
    cutsig  = sig[sigminloc:sigmaxloc]
    #print(cutsig)
    cutide  = ide[idemaxloc:ideminloc]
    #print(cutide)
    dataEnt = cutdata
    dataEnt_df = pd.DataFrame(cutdata, columns = cutsig, index = cutide)
    
    totEnt = 0
    for i in range(len(dataEnt)):
        
        for j in range(len(dataEnt[i])):
            totEnt += cutdata[i][j]
        
    Ssig = totEnt/totSig
    Nsig = (totSig-totEnt)/totSig
    Side = totEnt/totIde
    Nide = (totIde-totEnt)/totIde
    
    if plot == True:
        plt.figure(figsize=(10,7))
        ax = sns.heatmap(df)
        plt.xlabel("Signal, nm")
        plt.ylabel("Idler, nm")
        plt.show()
    
    return [Ssig, Nsig, Side, Nide]

def True_Success(total_distance,detector_number,source_rate,z1=0,z2=0,sigma=100*(10**9),ITU1=34,ITU2=34):

    PLoss = Pfibre(total_distance,detector_number)
    Omega1=ITU_dict[ITU1][1]
    Omega2=ITU_dict[ITU2][1]
    PSuccess = Pcoincidence_2bs(z1,z2,Omega1,Omega2,sigma)
    PTrue = PLoss*PSuccess*source_rate
    
    return PTrue

def False_Success(total_distance,detector_number,singles1,singles2,singles3,singles4,coincidence_window=250*(10**-12)):

    PLoss = Pfibre(total_distance,detector_number)
    PNoise = (singles1*singles2*singles3*singles4)*(coincidence_window**3)
    PFalse = PLoss*PNoise
    
    return PFalse

def False_SuccessNoBSM(total_distance,detector_number,singles1,singles2,coincidence_window=250*(10**-12)):

    PLoss = Pfibre(total_distance,detector_number)
    PNoise = (singles1*singles2)*coincidence_window
    PFalse = PLoss*PNoise
    
    return PFalse
```

```python
"""For a plot of Signal against individual fibre length"""
n = 4
source_rate = 10**6
S1 = 10*(10**6) #Hz
S2 = 10*(10**6) #Hz
S3 = 10*(10**6) #Hz
S4 = 10*(10**6) #Hz

TSs = []
FSs = []
BSM_N = []
BSM_S = []
dis = []
for i in np.arange(5,40.1,1):
    dtot = 4*i
    TS = True_Success(dtot,n,source_rate)
    FS = False_Success(dtot,n,S1,S2,S3,S4)
    BSM_Sig = (source_counts*Pfibre(2*i,2))*Pcoincidence_2bs(0,0)
    BSM_Noise = np.sqrt(((S1*Pfibre(2*i,2))**2)*Pcoincidence_2bs(0,0))
    dis.append(i)
    TSs.append(TS)
    FSs.append(FS)
    BSM_N.append(BSM_Noise)
    BSM_S.append(BSM_Sig)
    
TSs = np.array(TSs) 
FSs = np.array(FSs) 
BSM_S = np.array(BSM_S) 
BSM_N = np.array(BSM_N)
dis = np.array(dis)

plt.figure(figsize=(8,4))
plt.errorbar(dis,TSs,yerr=FSs,label="4-fold Coincidences",fmt='o-')
plt.errorbar(dis,BSM_S,yerr=BSM_N,label="Coincidences on Entanglement Swapping",fmt='o-')
plt.xlabel("Link length, km")
plt.ylabel("Coincidence Rate, 1/s")# (log10(1/s))")
plt.yscale('log')
plt.grid()

plt.legend()
plt.show()
    
```
